### Groq Model  intergration

```python
from langchain.chat_models import init_chat_model
model = init_chat_model(model="groq:qwen/qwen3-32b")
response= model.invoke("Why parrots talk?")
response
```

```python
from langchain_groq import ChatGroq
model= ChatGroq(model="qwen/qwen3-32b")
response = model.invoke("Why do parrots talk?")
response
```

```python


for chunk in model.stream("Write me 200 words paragraph on qunatum computing"):
    print(f"{chunk.text},", flush=True)
```

```python
responses=model.batch([
    "How do airplanes fly?", 
    "What is quantum computing?", 
    "Why do the parrots talk?"
], config={
    "max_concurrency":2
})

for response in responses:
    print(response)
```
